## 2.society

```python
import re

def clean_society_name(society):

    if isinstance(society, str):  # Ensure the input is a string

        # Remove "Sector" or "Sec" and everything following them
        cleaned = re.sub(r'\b(Sector|Sec)[^\s]*.*', '', society, flags=re.IGNORECASE)

        # Remove "Gurugram"
        cleaned = re.sub(r'\bgurugram\b', '', cleaned, flags=re.IGNORECASE)

        # Remove standalone numbers with optional trailing alphabets
        cleaned = re.sub(r'\d+[a-zA-Z]*\b', '', cleaned)

        # Remove hyphens
        cleaned = re.sub(r'-', '', cleaned)

        # Remove periods and stars
        cleaned = re.sub(r'[★.]', '', cleaned)

        return cleaned.strip()  # Remove leading/trailing spaces
    return society

# Apply the function to the DataFrame column

df['society'] = df['society'].apply(clean_society_name)
```

```python
def titleized(text):

    if isinstance(text, str):

      return str.title(text)

df['society'] = df['society'].apply(titleized)
```
